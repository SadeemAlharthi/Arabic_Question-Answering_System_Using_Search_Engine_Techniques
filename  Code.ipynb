```python
def preprocess(txt):
  for i in range(len(txt)):
    tokens = txt[i].split(' ')
    for j in range(len(tokens)):
      arr = [y for y in tokens[j]]
      arr.reverse()
      arr = ''.join(xx for xx in arr)
      tokens[j] = arr
    txt[i] = ' '.join(x for x in tokens)
  return txt
```

```python
def build_dict(data):
  dic = {}
  for i in range(len(data)):
    toks = data[i].split(' ')
    if toks[0] == 'المادة':
      key = data[i]
      val = []
      f = True
      m = 1 
      while f and (i+m) < len(data):
        tooks = data[i+m].split(' ')
        if tooks[0] != 'المادة':
          val.append(''.join(d for d in data[i+m]))
          m += 1
          f = True 
        else:
          f = False
        #print(val)
    dic[key] = ' ، '.join(m for m in val)
  return dic
```

```python
inverted_index = {}
for uid in t:
    for word in t[uid].split(' '):
        inverted_index.setdefault(word, {})[uid] = inverted_index.setdefault(word, {}).get(uid, 0) + 1
```

```python
df = {}
idf = {}
```

```python
import math
from collections import Counter
import operator
import webbrowser
```

```python
for key in inverted_index.keys():
    df[key] = len(inverted_index[key].keys())
    idf[key] = math.log(len(t.values()) / df[key], 2)
```

```python
df
```

```python
def tf_idf(w, doc):
    return inverted_index[w][doc] * idf[w]
```

```python
def inner_product_similarities(query):
    # dictionary in which I'll sum up the similarities of each word of the query with each document in
    # which the word is present, key is the doc number,
    # value is the similarity between query and document
    similarity = {}
    for w in str(query).split(' '):
        wq = idf.get(w, 0)
        if wq != 0:
            for doc in inverted_index[w].keys():
                similarity[doc] = similarity.get(doc, 0) + tf_idf(w, doc) * wq
    return similarity
```

```python
def doc_length(userid):
    words_accounted_for = []
    length = 0
    for w in t[userid].split(' '):
        if w not in words_accounted_for:
            length += tf_idf(w, userid) ** 2
            words_accounted_for.append(w)
    return math.sqrt(length)
```

```python
def query_length(query):
    length = 0
    cnt = Counter()
    for w in str(query).split(' '):
        cnt[w] += 1
    for w in cnt.keys():
        length += (cnt[w]*idf.get(w, 0)) ** 2
    return math.sqrt(length)
```

```python
def cosine_similarities(query):
    similarity = inner_product_similarities(query)
    
    for doc in similarity.keys():
        similarity[doc] = similarity[doc] / doc_length(doc)*query_length(query)
    return similarity
```

```python
def rank_docs(similarities):
    return sorted(similarities.items(), key=operator.itemgetter(1), reverse=True)
```

```python
from rouge import Rouge 
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.bleu_score import SmoothingFunction
smoothie = SmoothingFunction().method4
def new_query():
    query = input("أدخل البحث : ")
    if query is None:
        exit()
    query_tokens = query
    ranked_similarities = rank_docs(cosine_similarities(query_tokens))
    rouge = Rouge()
    scores = [rouge.get_scores(query, t[ranked_similarities[i][0]]) for i in range(len(ranked_similarities))]
    scores_bleu = [sentence_bleu(query, t[ranked_similarities[i][0]] , weights=(0.25, 0.25), 
                                 smoothing_function=smoothie) for i in range(len(ranked_similarities))]
  
    print('--------------------------------------')
    num_prints = input(" أدخل عدد المواد التي تريد طباعتها من النتيجة: ")
    if int(num_prints) > len(ranked_similarities):
      num_prints = input('لقد حدث خطأ فقد أدخلت رقم أكبر من نتائج البحث من فضلك أعد الإدخال مرة أخرى')
    print('--------------------------------------')
    print('')
    for i in range(int(num_prints)):
      print(ranked_similarities[i][0] + " : ",t[ranked_similarities[i][0]] )
      print('')
      print(" ROUGE Value = : ", scores[i])
      print(" BLEU Value = : ", scores_bleu[i])
```
